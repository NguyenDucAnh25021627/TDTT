```python
a, b = map(int, input().split())
def larger_num(a, b):
    return max(a, b)
print(larger_num(a, b))
```

```python
def swap (a, b):
    return b, a
a, b = map(int, input().split())
print(swap(a, b))
```

```python
def so_nguyen_to (n):
    for i in range (2, n // 2 + 1):
        if n % i == 0:
            return False
    return True
a = int(input())
print(so_nguyen_to(a))
```

```python
def so_hoan_hao (n):
    uoc = []
    for i in range (1, n):
        if n % i == 0:
            uoc.append(i)
    if sum(uoc) == n:
        return True
    else: 
        return False

print(so_hoan_hao(int(input())))
```

```python
def tim_vi_tri (l, n):
    i = 0
    for num in l:
        i += 1
        if num == n:
            return i
    return -1
print(tim_vi_tri(map(int, input().split()), int(input())))
```

```python
def giai_thua (n):
    giai_thua = 1
    for i in range(2, n + 1):
        giai_thua *= i
    return(giai_thua)
print(giai_thua(int(input())))
```

```python
def may_tinh (n):
    if n[1] == "+":
        return int(n[0]) + int(n[2])
    elif n[1] == "-":
        return int(n[0]) - int(n[2])
    elif n[1] == "/":
        return int(n[0]) / int(n[2])
    else:
        return int(n[0]) * int(n[2])

print(f"{may_tinh(input().split()):.2f}")
```

```python
def hamming_distance (a, b):
  c = a ^ b
  return bin(c).count('1')
a, b = map(int, input().split())
print(hamming_distance(a, b))
```

```python
def tong_chu_so (n):
  chu_so = []
  n = abs(n)
  for i in str(n):
    chu_so.append(int(i))
  return sum(chu_so)
print(tong_chu_so(int(input())))
```

```python
def don_cau (a,b):
    bang = {}
    i = 0
    for letter in a:
      anh_letter =  b[i]
      if letter in bang: 
        if bang[letter] != anh_letter:
          return False
      else:
        bang[letter] = anh_letter
      i += 1
    return True
def dang_cau (a,b):
    return don_cau(a,b) and don_cau(b,a)

a, b = input().split()
print(dang_cau(a, b))
```

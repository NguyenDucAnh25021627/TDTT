```python
def reverse_nunmber(n):
    t = []
    for i in str(n):
        t.append(i)
    t = list(reversed(t))
    temp_str = ""
    for digit in t:
        temp_str += digit
    return int(temp_str)

def is_palindrome(n):
    return n == reverse_nunmber(n)

n = int(input("Input n:"))
number_of_step = 0
while True:
    if not is_palindrome(n):
        number_of_step += 1
        n = n + reverse_nunmber(n)
    else:
        break
print(number_of_step,n)
```

```python
def find_collatz_len (n):
    collatz_len = 1
    while True:
        if n == 1:
            return collatz_len
        else:
            collatz_len += 1
            if n % 2 == 0:
                n = n / 2
            else:
                n = 3 * n + 1

n = int(input("Input n:"))

longest_len = 0
current_num = 0
for i in range (1,n + 1):
    if find_collatz_len(i) > longest_len:
        longest_len = find_collatz_len(i)
        current_num = i
print(current_num,longest_len)
```
